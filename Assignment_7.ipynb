Q2. Create a Python function called currency_converter() that takes two arguments:

- amount: The amount of money to be converted (in USD).
- exchange_rate (default is 280): The conversion rate from USD to another currency (e.g., PKR). You can assume the default rate is 1 USD = 280 PKR.
-
The function should return a message showing how much the given amount is worth in the target currency (PKR).

Expected Output:

'100 USD is equivalent to 28000 PKR'

```python
def currency_converter(amount, exchange_rate=280):
  return f'{amount} USD is equivalent to {amount*exchange_rate} PKR'

currency_converter(100)
```

Q5. Write a Python function calculate_rectangle that accepts the length and width of a rectangle as arguments. Inside this function, create two nested functions:

- calculate_area: This function should calculate the area of the rectangle.
- calculate_perimeter: This function should calculate the perimeter of the rectangle.
- The main function should return both the area and the perimeter of the rectangle.

Expected Output:

Area: 15

Perimeter: 16


```python
def calculate_rectangle(length,width):
  def calculate_area():
    return length*width
  def calculate_perimeter():
    return 2*(length+width)
  return calculate_area(),calculate_perimeter()

calculate_rectangle(3,5)
```

Q7:
Write a Python program that:

Has a balance of 1000.

Asks the user for a withdrawal amount.

Checks for these errors:

Non-numeric input → print "Please enter a valid number."

Amount ≤ 0 → print "Amount must be more than zero."

Amount > balance → print "Insufficient balance."

If valid, deducts amount and shows remaining balance.

Always prints "Thank you for using the ATM."

Expected Output:

- Enter amount to withdraw: 1500

  Insufficient balance.

  Thank you for using the ATM.


- Enter amount to withdraw:  0

  Amount must be more than zero.

  Thank you for using the ATM.


- Enter amount to withdraw:  K


  Please enter a valid number.

  Thank you for using the ATM.

```python
balance=1000
try:
  amount=float(input("Enter amount to withdraw: "))
  if amount<=0:
    print("Amount must be more than zero")
  elif amount > balance:
    print ("insufficient balance")
  else:
    balance-=amount
    print(f"Remaining balance is {balance}")
except ValueError:
  print("Please enter a valid number")
finally:
  print("Thank you for using the ATM")
```

Attributes: name, grade, and age.

A method introduce() that prints:

"My name is {name}, I am {age} years old."

A method show_grade() that prints:

"My grade is {grade}."

Create an object of the Student class with sample data (e.g., "Alice", "A", 16), and call both methods on the object.
